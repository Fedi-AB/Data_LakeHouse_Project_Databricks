## **Initialization**

## **Business Transformation and Modeling**

```python
query = """

SELECT

    ROW_NUMBER() OVER (ORDER BY pi.start_date, pi.product_number) AS product_key,
    pi.product_id,
    pi.product_number,
    pi.product_name,
    pi.product_line,
    pi.product_cost,
    pc.category_id,
    pc.category,
    pc.sub_category,
    pc.maintenance_flag,
    pi.start_date,
    pi.end_date
    
FROM 
    silver.crm_products pi
    LEFT JOIN 
        silver.erp_product_category pc
        ON pi.category_id = pc.category_id

"""

df = spark.sql(query)
```

## **Write Into Gold Layer**

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_products")
```
