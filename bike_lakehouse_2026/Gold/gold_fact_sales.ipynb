## **Initialization**

```python
import pyspark.sql.functions as F
from pyspark.sql.window import Window
```

## **Business Transformation and Modeling**

```python
query = """

SELECT

    cs.order_number,
    dp.product_key,
    dc.customer_key,
    cs.order_date,
    cs.ship_date,
    cs.due_date,
    cs.sales_amount,
    cs.quantity,
    cs.price
    
FROM 
    silver.crm_sales cs
    LEFT JOIN 
        gold.dim_products dp
        ON cs.product_number  = dp.product_number
    LEFT JOIN 
        gold.dim_customers dc
        ON cs.customer_id = dc.customer_id

"""

df = spark.sql(query)
```

## **Write Into Gold Layer**

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.fact_sales")
```
